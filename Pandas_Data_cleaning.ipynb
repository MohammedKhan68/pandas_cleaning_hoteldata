Date has text values also, so lets separate them in another column.

.isna function analyzes the column and gives T or F depnding on the avilaibility of na


```python
df['Room number'].isna()
```

```python
mask = df['Room number'].isna()
df['text value'] = np.where(mask, df['Date'], "")
df
```

The new column is created above, with the text value from the first column.
However, the empty cells are not good, lets add nan in place of empty.

```python
mask = df['Room number'].isna()
df['text value'] = np.where(mask, df['Date'], np.nan)
df
```

Lets fill the nan values.


```python
df['text value'].fillna(method='bfill', inplace=True)
```

```python
df
```

now lets drop nan values

```python
df.dropna(inplace=True)
df
```
